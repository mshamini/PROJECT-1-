#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def stdev(numbers):
    
    total = 0
    for i in numbers:
        total += i
        
# find overall mean
    x_mean = 0
    x_mean=total/len(numbers)
        
# x - x mean
    denom = 0
    sum_denom = 0
    for i in numbers:
        denom = (i-x_mean)**2
        sum_denom += denom
#St dev
    return np.sqrt( sum_denom/len(numbers))
````
